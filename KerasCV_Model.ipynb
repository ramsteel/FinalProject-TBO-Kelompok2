```python
classes = [
    'apple',
    'avocado',
    'cake',
    'cooked_chicken',
    'dragonfruit',
    'french_fries',
    'fried_rice',
    'hamburger',
    'noodles',
    'omelette',
    'salad'
]

class_mapping = dict(zip(range(len(classes)), classes))

# Path to images and annotations
IMAGES_DIR = "datasets/images/train"
LABELS_DIR = "datasets/labels/train"

# Get all XML file paths in path_annot and sort them
txt_files = sorted(
    [
        os.path.join(LABELS_DIR, file_name)
        for file_name in os.listdir(LABELS_DIR)
        if file_name.endswith(".txt")
    ]
)

# Get all JPEG image file paths in path_images and sort them
jpg_files = sorted(
    [
        os.path.join(IMAGES_DIR, file_name)
        for file_name in os.listdir(IMAGES_DIR)
        if file_name.endswith(".jpg")
    ]
)
```

```python
def parse_annotation(filename):
    boxes = []
    classes = []

    image_path = os.path.join(IMAGES_DIR, filename)

    with open(filename, 'r') as f:
      for label in f.read().splitlines():
          splitting_label = label.split()
          cls = int(splitting_label[0])
          classes.append(cls)

          x = float(splitting_label[1])
          y = float(splitting_label[2])
          w = float(splitting_label[3])
          h = float(splitting_label[4])
          boxes.append([x, y, w, h])

    return image_path, boxes, classes

image_paths = []
bbox = []
classes = []
for filename in tqdm(txt_files):
    image_path, boxes, class_ids = parse_annotation(filename)
    image_paths.append(image_path)
    bbox.append(boxes)
    classes.append(class_ids)
```

```python
a = tf.ragged.constant([1,2,3])
b = tf.ragged.constant([[1,1], [2,2], [3,3]])
c = tf.ragged.constant([[1,1], [2,2], [3,3]])

x = tf.data.Dataset.from_tensor_slices((a,b,c))
```

```python
bbox = tf.ragged.constant(bbox)
classes = tf.ragged.constant(classes)
image_paths = tf.ragged.constant(image_paths)

data = tf.data.Dataset.from_tensor_slices((image_paths, classes, bbox))
```

```python
# Determine the number of validation samples
num_val = int(len(txt_files) * SPLIT_RATIO)

# Split the dataset into train and validation sets
val_data = data.take(num_val)
train_data = data.skip(num_val)
```

```python
val_data
```

```python
def load_image(image_path):
    image = tf.io.read_file(image_path)
    image = tf.image.decode_jpeg(image, channels=3)
    return image


def load_dataset(image_path, classes, bbox):
    # Read Image
    image = load_image(image_path)
    bounding_boxes = {
        "classes": tf.cast(classes, dtype=tf.float32),
        "boxes": bbox,
    }
    return {"images": tf.cast(image, tf.float32), "bounding_boxes": bounding_boxes}
```

```python
train_ds = train_data.map(load_dataset, num_parallel_calls=tf.data.AUTOTUNE)
train_ds = train_ds.shuffle(BATCH_SIZE * 4)
train_ds = train_ds.ragged_batch(BATCH_SIZE, drop_remainder=True)
```

```python
val_ds = val_data.map(load_dataset, num_parallel_calls=tf.data.AUTOTUNE)
val_ds = val_ds.shuffle(BATCH_SIZE * 4)
val_ds = val_ds.ragged_batch(BATCH_SIZE, drop_remainder=True)
```
